# vizualizing sRNA (2nd idea)

```python
import pandas as pd
import json
from SPARQLWrapper import SPARQLWrapper, JSON
from pyvis.network import Network
```

```python
# url for query service
sparql = SPARQLWrapper('https://query.wikidata.org/sparql')
```

```python
# query
sparql.setQuery("""
SELECT DISTINCT ?srna ?srnaLabel ?target ?targetLabel
WHERE
{
 ?srna wdt:P31 wd:Q11053.
 ?srna wdt:P31 wd:Q427087.
 ?srna wdt:P128 ?target.
 
 SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}""")

sparql.setReturnFormat(JSON) # return as json
results = sparql.query().convert()
```

```python
# convert to pandas
query = pd.io.json.json_normalize(results['results']['bindings'])
```

```python
query
```

```python
# create empty graph with pyvis
net = Network(notebook=True)

# data
srna = query['srnaLabel.value']
target = query['targetLabel.value']
source = query['srna.value']

# zip data
edge_data = zip(srna, target, source)

for e in edge_data:
    srna = e[0]        
    target = e[1]
    src = e[2]
    
    #add nodes and edges
    net.add_node(srna, title=srna) 
    net.add_node(target, title=target) 
    net.add_edge(srna, target, title=src)
```

```python
net.show('test.html')
```
