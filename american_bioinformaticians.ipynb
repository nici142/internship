# wikidata queries with python

## get query as json file and visualize it in networkx
### example: all bioinformaticians with the nationality USA

```python
import requests # HTTP library for Python
import networkx as nx
import matplotlib.pyplot as plt

url = 'https://query.wikidata.org/sparql'

# query from sparql
query = """
SELECT ?item ?itemLabel                          
WHERE 
{
  ?item wdt:P31 wd:Q5.
  ?item wdt:P106 wd:Q2904006.
  ?item wdt:P27 wd:Q30.
  
  SERVICE wikibase:label { bd:serviceParam wikibase:language "[AUTO_LANGUAGE],en". }
}"""

r = requests.get(url, params={'format': 'json', 'query': query}) # get content as json
data = r.json() # returns data as json
```

```python
bioinformaticians = [] # empty list
                
                    # keys
for item in data['results']['bindings']:
    bioinformaticians.append(item['itemLabel']['value'])
```

```python
print(bioinformaticians) # list with names of all bioinformaticians with citizenship in the USA
```

```python
G = nx.Graph() # create empty graph
```

```python
# add list as edges
for edges in bioinformaticians:
    G.add_edge("bioinformaticians", edges)
```

```python
# show all american bioinformaticians in a graph
nx.draw(G)
plt.show()
```
