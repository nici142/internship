# json file in ein python object machen

```python
import urllib.request
import json

url= urllib.request.urlopen("http://api.crossref.org/works/10.1128/JB.00730-18").read() # opens and reads url
data=json.loads(url) #loads übersetzt string in ein dict
print(data)
```

# do more json files to python object

```python
import urllib.request
import json

url= "http://api.crossref.org/works/"

dois=["10.1128/JB.00730-18","10.1371/journal.pbio.3000057"] # list of dois

for dois in dois: # for schleife geht alles in der liste durch
    result = urllib.request.urlopen(url+dois).read()
    data=json.loads(result)
    print(data)
    print("----------------------------")
```
